```python
# Frequent Itemsets Mining
```

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('./Book_Fre.csv',index_col='TID')
df.head(6)
```

```python
Frequent_itemset = apriori(df, min_support=0.5, use_colnames = True)
Frequent_itemset
```

```python
Frequent_itemset.to_csv('Book_FI.csv')
```

```python
A_R = association_rules(Frequent_itemset, metric = "lift", min_threshold = 1)
A_R.sort_values('lift',ascending = False)
A_R
```

```python
A_R.to_csv('Book_AR.csv')
```
