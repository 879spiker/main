```python
# seperate data as training and testing
#   test_size: 0.2~0.3
#   shuffle = True (random sequence)
#   veiw shape of training data and test data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
scaled_data = min_max_scaler.fit_transform(boston.data)

X_train, X_test, y_train, y_test = train_test_split(df.drop('PRICE', axis=1), boston.target, test_size=0.3, random_state= 12, shuffle=True)
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression # select model
model = LinearRegression() # build model
model.fit(X_train,y_train) # train model
training = model.predict(X_train)
y_test_pred = model.predict(X_test) # model prediction
```

```python
# model Evaluation
# by R2 score
```

```python
from sklearn import metrics
print('R2 score for training set:%.2f'%metrics.r2_score(y_train, training))
print('R2 score for accuracy:%.2f'%metrics.r2_score(y_test, y_test_pred))
```
