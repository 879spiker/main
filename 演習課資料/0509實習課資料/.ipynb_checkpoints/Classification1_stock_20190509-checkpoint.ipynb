```python
# Classification for Stock: KNN, Bayes, Decision Tree
```

```python
# import package and datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics, preprocessing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('2882_17v1.csv')
```

```python
df.head()
```

```python
# normalize
def normalize(df):
    norm = df.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
    return norm
norm = normalize(df[['開盤','最高','最低','收盤','成交值千元','外資買賣超金額','投信買賣超金額','自營商買賣超金額','融資餘額張數','融券餘額張數']])
```

```python
norm.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(norm, df['1d'], test_size=0.3)
```

```python
# check the training and testing data~
print(len(X_test))
print(len(X_train))
print(len(y_test))
print(len(y_train))
```

```python
# KNN algorithm

# import algorithm from sklearn
from sklearn import neighbors

# build model
model2 = neighbors.KNeighborsClassifier(n_neighbors=5)

# training data
model2.fit(X_train, y_train) 

# predict data
y_pred=model2.predict(X_test)

# print result
print(y_pred)
print(y_test)
```

```python
# print report
print(metrics.confusion_matrix(y_test,y_pred))
print(metrics.accuracy_score(y_test,y_pred))
print(metrics.classification_report(y_pred,y_test))
```

```python
# Naive Bayes Algorithm
```

```python
# Naive Bayes

# from sklearn.naive_bayes import GaussianNB
from sklearn import naive_bayes

# build model
model4 = naive_bayes.GaussianNB()
#kernel要調的好, 但會有過度學習的狀況

# training data
model4.fit(X_train, y_train) 

# predict
y_pred=model4.predict(X_test)

# print result
print(y_pred)
print(y_test)
```

```python
# print report
print(metrics.confusion_matrix(y_test,y_pred))
print(metrics.accuracy_score(y_test,y_pred))
print(metrics.classification_report(y_pred,y_test))
```

```python
# Decision Tree Algorithm
```

```python
# decision tree
from sklearn import tree

# build model
model1 = tree.DecisionTreeClassifier()

# training data
model1.fit(X_train, y_train) 

# predict data
y_pred=model1.predict(X_test)

# print reulst
print(y_pred)
print(y_test)
```

```python
# print report
print(metrics.confusion_matrix(y_test,y_pred))
print(metrics.accuracy_score(y_test,y_pred))
print(metrics.classification_report(y_pred,y_test))
```
