# 建立 pandas.Series

Pandas Series 是一個擁有標籤的一維陣列，可以將 Series 視為 2 個陣列的組合，一個是類似索引（index）的標籤，另一個是實際資料。可以使用 Array、Dict、Scalar value or constant 來建立

這結構與 dict 字典類似，字典是一種無序的數據結構，而 Series 是不一樣，是相當於定長有序的字典，且它的 index 和 value 之間是獨立的。兩者的索引還是有區别的，Series 的 index 是可變的，而 dict 字典的 key 值是不可變的

**pandas.Series(data = None, index = None, dtype = None, name = None, copy = False, fastpath = False)**

參數描述：
- data：數據採用各種形式，如 ndarray，list，常量
- index：索引值必須是唯一且可清除的，與數據長度相同。如果沒有傳遞索引，則預設為 np.arrange(n)
- dtype：dtype 用於數據類型。如果為 None，則將推斷數據類型
- copy：複製數據。預設為 False

## 丟棄缺失值

有缺失值，常見的一種處理辦法就是丟棄缺失值

使用 dropna 方法可以丟棄缺失值

```python
user_info.age.dropna()
```

Seriese 使用 dropna 較簡單，對於 DataFrame 有更多的參數可以使用

- axis 参數用於控制行或列，axis=0 （預設）表示操作列，axis=1 表示操作行

- how 参數可選的值為 any（預設） 或 all。any 表示一行/列有任意元素為空時即丢棄，all 一行/列所有值都為空時才丟棄

- subset 参數表示删除時只考慮的索引或列名

- thresh 参數的類型為整數，它的作用是，比如 thresh=3，會在一行/列中至少有 3 個非空值時將其保留

```python
# 一列數據只要有一個字段存在空值即删除
user_info.dropna(axis = 0, how = 'any')
```

```python
# 一欄數據只要有一個字段存在空值即删除
user_info.dropna(axis = 1, how = 'any')
```

```python
# 一列數據所有字段都為空值才删除
user_info.dropna(axis = 0, how= 'all')
```

```python
user_info.dropna(axis = 1, how = 'all')
```

```python
# 一列數據中只要 city 或 sex 欄存在空值即删除
user_info.dropna(axis = 0, how = 'any',  subset = ['city', 'sex'])
```

## 聚合

分組的目的是為了統計，統計的時候需要聚合，常見的一些聚合操作有：計數、求和、最大值、最小值、平均值等

想要實現聚合操作，一種方式就是調用 agg 方法

```python
# 取得不同性别下所包含的人數
grouped = user_info.groupby('sex')
grouped['age'].agg(len)
```

```python
# 取得不同性别下包含的最大的年齡
grouped = user_info.groupby('sex')
grouped['age'].agg(np.max)
```

如果是根據多個鍵來進行聚合，預設情況下得到的結果是一個多層索引結構

```python
grouped = user_info.groupby(['sex', 'age'])
rs = grouped.agg(len)
rs
```

有兩種方式可以避免出現多層索引

第一種方法是對包含多層索引的對象使用 reset_index()

```python
rs.reset_index()
```

另一種方式是在分組時，設置参數 as_index=False

```python
grouped = user_info.groupby(['sex', 'age'], as_index = False)
grouped.agg(len)
```

分組後一樣可以使用 describe 方法來查看數據的情況

```python
grouped = user_info.groupby('sex')
grouped.describe()
```

## 一次應用多個聚合操作

進行分組後，不單單想得到一個統計結果，有可能是多個

比如，想統計出不同性别下的一個收入的總和和平均值

```python
grouped = user_info.groupby('sex')
grouped['income'].agg([np.sum, np.mean])
```

要將統計結果進行重命名，可以傳入字典格式

```python
grouped = user_info.groupby('sex')
grouped['income'].agg([np.sum, np.mean]).rename(columns = {'sum':'income_sum', 'mean':'income_mean'})
```

## 對 DataFrame 行應用不同的聚合操作

需要對不同的行使用不同的聚合操作

例如，想要統計不同性别下人群的年齡的均值以及收入的總和

```python
grouped = user_info.groupby('sex')
grouped.agg({'age':np.mean, 'income':np.sum}).rename(columns = {'age':'age_mean', 'income':'income_mean'})
```

## transform 操作

前面進行聚合運算時，得到的結果是一個以分組名作為索引的結果對象，雖然可以指定 as_index=False ，但是得到的索引也並不是原數據的索引，如果想使用原數據的索引的话，就需要進行 merge 轉換

transform 方法簡化了這個過程，會把 func 參數應用到所有分組，然後把結果放置到原數據的索引上（如果結果是一個標量，就進行廣播）

```python
# 透過 agg 得到的結果的索引是分組名
grouped = user_info.groupby('sex')
grouped['income'].agg(np.mean)
```

```python
# 透過 transform 得到的結果的索引是原始索引，會將得到的結果自動關聯上原始的索引
grouped = user_info.groupby('sex')
grouped['income'].transform(np.mean)
```

可以看到，透過 transform 操作得到結果的長度與原來保持一致

# 關聯

有兩個 DataFrame，分别用戶的部分資訊，現在需要將這些資訊關聯起来，有以下方法可以實現

```python
data1 = {
    'name':['Tom', 'Bob', 'Mary', 'James'], 
    'age':[18, 30, 35, 18], 
    'city':['Bei Jing', 'Shang Hai', 'Guang Zhou', 'Shen Zhen']
}
```

```python
df1 = pd.DataFrame(data = data1)
df1
```

```python
data2 = {
    'name':['Bob', 'Mary', 'James', 'Andy'], 
    'sex':['male', 'female', 'male', np.nan], 
    'income':[8000, 8000, 4000, 6000]
}
```

```python
df2 = pd.DataFrame(data = data2)
df2
```

## merge

透過 pd.merge() 可以關聯兩個 DataFrame，設置參數 on='name'，表示依據 name 來作為關聯鍵

```python
pd.merge(df1, df2, on = 'name')
```

關聯後發現數據變少了，只有 3 列數據，這是因為預設關聯的方式是 inner，如果不想丢失任何數據，可以設置參數 how='outer'

```python
pd.merge(df1, df2, on = 'name', how = 'outer')
```

在設置 how='outer' 後，就不會遺漏任何數據，也會在不存在的地方填入缺失值

如果想保留左邊所有的數據，可以設置參數 how='left'；反之，如果想保留右邊的所有數據，可以設置參數 how='right'

```python
pd.merge(df1, df2, on = 'name', how = 'left')
```

當兩個 DataFrame 中需要關聯的鍵的名稱不一樣，可以透過 left_on 和 right_on 來分別設置

```python
df1.rename(columns = {'name': 'name1'}, inplace = True)
df1
```

```python
df2.rename(columns = {'name': 'name2'}, inplace = True)
df2
```

```python
pd.merge(df1, df2, left_on = 'name1', right_on = 'name2')
```

有時候，兩個 DataFrame 中都包含相同名稱的字段，可以設置參數 suffixes，預設 suffixes=('_x', '_y') 表示將相同名稱的左邊的 DataFrame 的字段名加上後缀 _x，右邊加上後缀 _y

```python
df1['sex'] = 'male'
df1
```

```python
pd.merge(df1, df2, left_on = 'name1', right_on = 'name2')
```

```python
pd.merge(df1, df2, left_on = 'name1', right_on = 'name2', suffixes = ('_left', '_right'))
```

## join

除了 merge 這種方式外，還可以透過 join 這種方式實現關聯。相比 merge，join 這種方式有以下幾個不同：

預設參數 on=None，表示關聯時使用左邊和右邊的索引作為鍵，設置參數on可以指定的是關聯時左邊的所用到的鍵名

左邊和右邊字段名稱重複時，透過設置參數 lsuffix 和 rsuffix 來解決

```python
df1.join(df2.set_index('name2'), on = 'name1', lsuffix = '_left')
```
