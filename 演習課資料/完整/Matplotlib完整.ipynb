Matplotlib 是建立在 NumPy 上一個處理資訊視覺化的程式庫

最重要的特色是，可以在不同的作業系統和繪圖後端執行。雖然近幾年 Matplotlib 的介面相較於其他如 R 語言的 ggplot、ggvis 具有網頁視覺化的工具包，顯得老態，但後人也在其強大的內在中開發了新的套件去驅動 Matplotlib 更現代的 API，如 Seaborn、ggplot、Altair，甚至是 Pandas 也包裝了Matplotlib 的 API，直接在 Pandas 中進行簡易繪圖

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 使用 2 個資料集繪製 2 條折線

```python
x = np.linspace(0, 10, 50)    # numpy.linspace(start, stop, num)    num 樣本數量
sinus = np.sin(x)
cosinus = np.cos(x)
```

```python
plt.plot(x, sinus, x, cosinus)

plt.show()
```

# 更改線條的外觀

```python
plt.plot(x, sinus, 'r-o', x, cosinus, 'g--')

plt.show()
```

# 顯示圖表的格線

```python
plt.plot(x, sinus, 'r-o', x, cosinus, 'g--')
plt.grid(True)

plt.show()
```

# 顯示 x 和 y 軸的座標軸標籤

```python
plt.plot(days, celsius)
plt.xlabel('Day')
plt.ylabel('Celsius')

plt.show()
```

# 顯示圖表的標題文字

```python
plt.plot(x, sinus, 'r-o', x, cosinus, 'g--')
plt.xlabel('Rade')
plt.ylabel('Amplitude')
plt.title('Sin and Cos Waves')

plt.show()
```

# 顯示圖表的圖例

```python
plt.plot(x, sinus, 'r-o', label = 'sin(x)')
plt.plot(x, cosinus, 'g--', label = 'cos(x)')
plt.legend()
plt.xlabel('Rade')
plt.ylabel('Amplitude')
plt.title('Sin and Cos Waves')

plt.show()
```

```python
plt.plot(x, sinus, 'r-o', label = 'sin(x)')
plt.plot(x, cosinus, 'g--', label = 'cos(x)')
plt.legend(loc = 1)     # 'upper right'      值從 0~10
plt.xlabel('Rade')
plt.ylabel('Amplitude')
plt.title('Sin and Cos Waves')

plt.show()
```

# 指定多個資料集的軸範圍

```python
days = range(1, 9)
celsius_min = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1]
celsius_max = [27.6, 26.1, 22.5, 30.4, 29.5, 31.5, 35.1, 39.4]
```

```python
plt.plot(days, celsius_min, 'r-o', days, celsius_max, 'g--o')
plt.xlabel('Day')
plt.ylabel('Celsius')
plt.axis([0, 10, 15, 40])

plt.show()
```

# 儲存圖表

```python
plt.plot(days, celsius_min, 'r-o', days, celsius_max, 'g--o')
plt.xlabel('Day')
plt.ylabel('Celsius')
plt.axis([0, 10, 15, 40])
plt.savefig('./Celsius.png')

plt.show()
```

```python
plt.savefig(filename = './Celsius', format = 'png')   # filename 是檔名，format 是檔案格式
```

# 繪製 2 張垂直排列的子圖表

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
consinus = np.cos(x)
```

```python
plt.subplot(2, 1, 1)     # 繪製在 2x1 表格（前兩個參數）的第 1 列（第三個參數）
plt.plot(x, sinus, 'r-o')
plt.subplot(2, 1, 2)    # 繪製在 2x1 表格的第 2 列
plt.plot(x, consinus, 'g--')

plt.show()
```

# 繪製 2 張水平排列的子圖表

```python
plt.subplot(1, 2, 1)      # 在 1x2 表特的第 1 欄
plt.plot(x, sinus, 'r-o')
plt.subplot(1, 2, 2)    # 在 1x2 表特的第 2 欄
plt.plot(x, consinus, 'g--')

plt.show()
```

# 繪製 6 張表格排列的子圖表

```python
plt.subplot(231)     
plt.plot(x, np.sin(x))
plt.subplot(232)   
plt.plot(x, np.cos(x))
plt.subplot(233)
plt.plot(x, np.tan(x))
plt.subplot(234) 
plt.plot(x, np.sinh(x))
plt.subplot(235)
plt.plot(x, np.cosh(x))
plt.subplot(236)
plt.plot(x, np.tanh(x))

plt.show()
```

subplot() 的 3 個參數值都小於 10，可以使用 1 個整數值的參數代替 3 個整數值的參數值，原來第 1 個參數值是百進位值，第 2 個參數值是十進位值，最後是個位值

# 繪製多軸圖表共用 x 軸的 2 個資料集

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
sinhs = np.sinh(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, sinus, 'r-o')
ax2 = ax.twinx()
ax2.plot(x, sinhs, 'g--')

plt.show()
```

# 顯示共用 x 軸多軸圖表的標籤文字

```python
fig, ax = plt.subplots()
ax.plot(x, sinus, 'r-o')
ax.set_xlabel('x', color = 'green')
ax.set_ylabel('Sin(x)', color = 'red')
ax2 = ax.twinx()
ax2.plot(x, sinhs, 'g--')
ax2.set_ylabel('Sinh(x)', color = 'blue')

plt.show()
```

# 顯示共用 x 軸多軸圖表的圖例

```python
fig, ax = plt.subplots()
ax.plot(x, sinus, 'r-o', label = 'Sin(x)')
ax.set_xlabel('x', color = 'green')
ax.set_ylabel('Sin(x)', color = 'red')
ax.legend(loc = 'best')
ax2 = ax.twinx()
ax2.plot(x, sinhs, 'g--', label = 'Sinh(x)')
ax2.set_ylabel('Sinh(x)', color = 'blue')
ax2.legend(loc = 'best')

plt.show()
```

# 在共用 x 軸多軸圖表顯示單一圖例

```python
fig, ax = plt.subplots()
lns1 = ax.plot(x, sinus, 'r-o', label = 'Sin(x)')
ax.set_xlabel('x', color = 'green')
ax.set_ylabel('Sin(x)', color = 'red')
ax2 = ax.twinx()
lns2 = ax2.plot(x, sinhs, 'g--', label = 'Sinh(x)')
ax2.set_ylabel('Sinh(x)', color = 'blue')
# 自行建立圖例來顯示所有標籤
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc = 'best')

plt.show()
```

# 更改圖表的外觀

```python
fig, ax = plt.subplots()
ax.plot(x, sinus, 'r-o')
ax.set_xlabel('x', color = 'green')
ax.set_ylabel('Sin(x)', color = 'red')
ax2 = ax.twinx()
ax2.plot(x, sinhs, 'g--')
ax2.set_ylabel('Sinh(x)', color = 'blue')

# 指定圖表標題文字
ax.set_title('Sin and Sinh Waves', fontsize = 'large')

# 更改刻度的外觀
for tick in ax.xaxis.get_ticklabels():
    tick.set_fontsize('large')
    tick.set_fontname('Times New Roman')
    tick.set_color('blue')
    tick.set_weight('bold')

plt.show()
```
